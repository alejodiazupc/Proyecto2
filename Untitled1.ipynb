Pregunta 1: Extracción de datos de acciones de Tesla utilizando yfinance

```python
import yfinance as yf
```

```python
# Crear objeto ticker para Tesla
tesla = yf.Ticker("TSLA")
```

```python
# Obtener datos históricos de acciones
tesla_data = tesla.history(period="max")
```

```python
# Reiniciar índice y mostrar las primeras 5 filas
tesla_data.reset_index(inplace=True)
print(tesla_data.head())
```

```python
print(tesla_data.tail())
```

Pregunta 2: Extracción de datos de ingresos de Tesla utilizando Webscraping






```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


# URL de los ingresos de Tesla
url = "https://es.investing.com/equities/tesla-motors-income-statement"
response = requests.get(url)
```

```python

# Creating BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Encuentra todas las tablas que tienen las clases "genTbl" y "reportTbl"
tablas = soup.find_all('table', {'class': lambda class_value: 'genTbl' in class_value and 'reportTbl' in class_value})
```

```python
# Extraer los datos de la tabla
# Crear una lista para almacenar los datos de la tabla
data = []

# Iterar sobre cada tabla encontrada
for tabla in tablas:
    # Encontrar todas las filas de la tabla
    filas = tabla.find_all('tr')

    # Iterar sobre cada fila
    for fila in filas:
        # Encontrar todas las celdas de la fila
        celdas = fila.find_all(['th', 'td'])

        # Verificar si la fila tiene 5 celdas
        if len(celdas) == 5:
            # Extraer datos de las celdas y agregarlos a la lista de datos
            fila_datos = [celda.get_text(strip=True) for celda in celdas]
            data.append(fila_datos)
# Imprimir los datos extraídos
i = 0
for row in data:
    print(str(i), row)
    i += 1
```

```python
# Convertir a DataFrame de pandas
tesla_revenue = pd.DataFrame(data)
# Mostrar el DataFrame
print(tesla_revenue.tail())
```

Pregunta 5: Tablero de acciones de Tesla

```python
import matplotlib.pyplot as plt
import pandas as pd

# Graficar el precio de cierre a través del tiempo
plt.figure(figsize=(12, 6))
plt.plot(tesla_data['Date'], tesla_data['Close'])
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre')
plt.title('Tesla - Precio de Cierre')
plt.xticks(rotation=45)
plt.show()

# Graficar el volumen a través del tiempo
plt.figure(figsize=(12, 6))
plt.bar(tesla_data['Date'], tesla_data['Volume'])
plt.xlabel('Fecha')
plt.ylabel('Volumen')
plt.title('Tesla - Volumen de Transacciones')
plt.xticks(rotation=45)
plt.show()

# Graficar precios de apertura, máximo, mínimo y cierre
plt.figure(figsize=(12, 6))
plt.plot(tesla_data ['Date'], tesla_data ['Open'], label='Apertura')
plt.plot(tesla_data ['Date'], tesla_data ['High'], label='Máximo')
plt.plot(tesla_data ['Date'], tesla_data ['Low'], label='Mínimo')
plt.plot(tesla_data ['Date'], tesla_data ['Close'], label='Cierre')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.title('Tesla - Precios Diarios')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
